```python
def plot_close_val(data_frame, column, stock,company):
    plt.figure(figsize=(16,6))
    dft=df.loc[data_frame[company]==stock]
    
    plt.title(column + ' Price History for '+ str(stock) )
    plt.plot(dft[column])
    plt.xlabel('Date', fontsize=18)
    plt.ylabel(column + ' Price USD ($) for ' + str(stock), fontsize=18)
    plt.show()



```

```python
#Test the function
plot_close_val(df, 'Close', 1,'Company')
plot_close_val(df, 'Close', 2, 'Company')
plot_close_val(df, 'Close', 3, 'Company')
plot_close_val(df, 'Close', 4, 'Company')
```

```python
def build_training_dataset(input_ds):
    # Create a new dataframe with only the 'Close column 
    input_ds.reset_index()
    data = input_ds.filter(items=['Close'])
    # Convert the dataframe to a numpy array
    dataset = data.values
    # Get the number of rows to train the model on
    training_data_len = int(np.ceil( len(dataset) * .95 ))
    return data, dataset, training_data_len
```
